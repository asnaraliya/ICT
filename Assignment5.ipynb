# Exploratory Data Analysis with Seaborn
This notebook explores three built-in datasets (Flights, Tips, Penguins) using Seaborn visualizations. Each plot is accompanied by insights to highlight patterns and relationships in the data.


```python
# Step 1: Install seaborn if not already installed
!pip install seaborn

# Step 2: Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for better visuals
sns.set(style="whitegrid")
```

```python
# Load Flights dataset
flights = sns.load_dataset("flights")

# Pivot for heatmap
pivot = flights.pivot(index="month", columns="year", values="passengers")

# Plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap="YlGnBu")
plt.title("Monthly Airline Passengers (1949–1960)")
plt.show()
```

```python
# Load Tips dataset
tips = sns.load_dataset("tips")

# Scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(x="total_bill", y="tip", hue="time", size="size", data=tips)
plt.title("Tips vs Total Bill")
plt.show()
```

```python
# Load Penguins dataset
penguins = sns.load_dataset("penguins")

# Countplot of species by island
plt.figure(figsize=(8,6))
sns.countplot(x="island", hue="species", data=penguins)
plt.title("Penguin Species Distribution by Island")
plt.show()
```

```python
# Correlation heatmap for numerical features
plt.figure(figsize=(8,6))
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Penguin Features")
plt.show()
```
