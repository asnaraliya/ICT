## Challenge 1: Word Frequency Analyzer
Write a function `top_k_words(filename: str, k: int) -> list` that returns the top k frequent words.
- Ignore case and punctuation.
- Sort by frequency, then alphabetically.


```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    # Read the file content
    with open(filename, 'r') as file:
        content = file.read()
    # Extract words , ignore case and punctuation
    words = re.findall(r'\b\w+\b' , content.lower())

    # Count the frequency of each word
    counts = Counter(words)

    # Sort by frequency (descending),then alphabetical (ascending)
    # -X[1] handles the descending order for frequency
    # X[0] handles the ascending order for alphabetical words
    sorted_words = sorted(counts.items(), key=lambda x: (-x[1], x[0]))

    #Extract just the words for the top k items
    return [word for word,count in sorted_words[:k]]




# Sample Test (create a sample file)
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))  #expected ["is", "data", "science"]
```

## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

        # TODO: store category-wise total
        if category in expenses:
            expenses[category] +=amount
        else:
           expenses[category] =amount




    # TODO: calculate total expense
    total = sum(expenses.values())

    # TODO: find highest category
    # max() returns the key with the highest value using expenses.get
    highest_category = max(expenses , key=expenses.get) if expenses else None

    print("Total Expense:" ,total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)

analyze_expenses()
```
