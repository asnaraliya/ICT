```python
Months = ['Jan', 'Feb', 'March', 'April', 'May', 'June']
visits = [12500,14200,16800,21000,19700,30100]
conversion_rates = [2.5,3.1,2.8,3.5,3.8,4.8]
converted_visits = np.array(visits) * np.array(conversion_rates) / 100
avg_visits = np.mean(visits)
plt.plot(Months,visits,'o-' ,color ='red')
plt.xlabel('Months')
plt.ylabel('Visits')
plt.title('Monthly Visits')
plt.plot(Months,converted_visits,'D-' ,color ='green')
plt.legend(['Visits','Converted Visits','Converted visits'])
plt.axhline(y=avg_visits, color='blue', linestyle='--')
```

```python
plt.bar(Months,visits,color='blue',edgecolor="black",width=0.4)
plt.xlabel('Months')
plt.ylabel('Visits')
plt.title('Monthly Visits')
plt.bar(Months,converted_visits,color='red')
plt.legend(['Visits','Converted Visits'])
```
