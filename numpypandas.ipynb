```python
import numpy as np

arr = np.arange(10, 50)
matrix = arr.reshape(4, 10)
last_col = matrix[:, -1]

print(matrix)
print("Last column:", last_col)
```

```python
A = np.full((3, 3), 5)
B = np.arange(1, 10).reshape(3, 3)

add = A + B
multiply = A * B
transpose = multiply.T

print("Addition:\n", add)
print("Multiplication:\n", multiply)
print("Transpose:\n", transpose)
```

```python
rand_arr = np.random.randint(10, 100, 15)
replaced = np.where(rand_arr > 50, -1, rand_arr)
count = np.sum(rand_arr > 50)

print("Original:", rand_arr)
print("Replaced:", replaced)
print("Count replaced:", count)
```

```python
arr = np.random.rand(5, 4)
row_sums = arr.sum(axis=1)
max_row_index = np.argmax(row_sums)

normalized = (arr - arr.min(axis=1)[:, None]) / (arr.max(axis=1) - arr.min(axis=1))[:, None]

print("Array:\n", arr)
print("Row with max sum:", max_row_index)
print("Normalized:\n", normalized)
```

```python
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique = np.unique(arr)
duplicates = [x for x in unique if np.count_nonzero(arr == x) > 1]

print("Unique:", unique)
print("Duplicates:", duplicates)
```

```python
arr = np.arange(1, 37).reshape(6, 6)
submatrix = arr[1:5, 1:5]
evens = submatrix[submatrix % 2 == 0]

print("Submatrix:\n", submatrix)
print("Even numbers:", evens)
```

```python
arr = np.arange(12).reshape(3, 4)
reshaped = arr.reshape(2, 6)
flattened = reshaped.flatten()
back = flattened.reshape(4, 3)

print("Reshaped:\n", reshaped)
print("Flattened:", flattened)
print("Back to (4,3):\n", back)
```

```python
a = np.array([[1, 2], [3, 4]])
a = np.vstack([a, [5, 6]])
a = np.column_stack([a, [7, 8, 9]])

print(a)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])

joined = np.concatenate([a, b, c]).reshape(3, 3)
diag = np.diag(joined)

print("Matrix:\n", joined)
print("Diagonal:", diag)
```

```python
arr = np.arange(1, 21)
divisible = arr[(arr % 3 == 0) | (arr % 5 == 0)]
count = divisible.size

print("Divisible:", divisible)
print("Count:", count)
```
