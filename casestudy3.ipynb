```python
import pandas as pd
import numpy as np

print("Libraries imported successfully!")
```

```python
df = pd.read_csv("/content/IMDB-Movie-Data.csv")
print("Dataset loaded successfully!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
print(df.head())
print(df.dtypes)
print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

```python
df.rename(columns={"Title":"Movie_Title"}, inplace=True)
print(df.columns)
```

```python
print(df['Genre'].unique()[:10])
```

```python
df['Num_Genres'] = df['Genre'].apply(lambda x: len(x.split(',')))
print(df[['Movie_Title','Genre','Num_Genres']].head())
```

```python
print("Missing before filling:\n", df.isnull().sum())
df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].median(), inplace=True)
df['Metascore'].fillna(df['Metascore'].median(), inplace=True)
print("Missing after filling:\n", df.isnull().sum())
```

```python
print(df.nlargest(5,'Rating')[['Movie_Title','Rating']])
```

```python
print("Highest Revenue Movie:\n", df.loc[df['Revenue (Millions)'].idxmax()][['Movie_Title','Revenue (Millions)']])
print("\nLowest Revenue Movie:\n", df.loc[df['Revenue (Millions)'].idxmin()][['Movie_Title','Revenue (Millions)']])
```

```python
print(df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10))
```

```python
print(df['Year'].value_counts().sort_index())
```

```python
print(df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)][['Movie_Title','Rating','Revenue (Millions)']])
```

```python
print(df.groupby('Director')[['Rating','Revenue (Millions)']].mean().head())
```

```python
print(df.groupby('Year').agg({'Movie_Title':'count','Rating':'mean'}))
```

```python
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
print(df.nlargest(10,'Profitability')[['Movie_Title','Revenue (Millions)','Runtime (Minutes)','Profitability']])
```

```python
def categorize_rating(r):
    if r < 6: return "Poor"
    elif 6 <= r < 7.5: return "Average"
    else: return "Excellent"

df['Rating_Category'] = df['Rating'].apply(categorize_rating)
print(df[['Movie_Title','Rating','Rating_Category']].head())
```

```python
def high_score(row):
    return row['Rating']*10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(high_score, axis=1)
print(df[['Movie_Title','High_Score']].head())
```
