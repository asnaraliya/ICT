```python
# Variables and data types for accoount information
account_holder = "Anjali"
balance = 5000.0

# Functions for bank operations
def balance_enquiry():
    print(f"Account Holder:{account_holder}")
    print(f"current Balance:INR{balance:.2f}")

def deposit(amount):
    global balance
    if amount > 0:
       balance += amount
       print(f"Deposited:{amount:.2f} | New Balance:INR {balance:.2f}")
    else:
      print("Invalid deposit amount.")

def withdrawal(amount):
   global balance
   # Conditional statement to check for sufficient balance
   if amount > balance:
       print(f"Withdrawal Denied:Insufficient balance.Available: {balance:.2f}")
   elif amount <= 0:
       print("Invalid withdrawal amount.")
   else:
       balance -= amount
       print(f"withdrawn: INR{amount:.2f} | Remaining Balance:INR{balance:.2f}")

 # Stimulation execution
   print("---Initial Account State ---")
   balance_enquiry()

print("\n---Testing Deposit ---")
deposit(1500.0)

print("\n---Testing Valid Withdrawal ---")
withdrawal(2000.0)

print("\n---Testing Insufficient Balance Check ---")
withdrawal(6000.0)

print("\n---Final Balance Enquiry ---")
balance_enquiry()


```

```python
# --- 2b. Variables and data types to store weather data ---
# We use lists of numbers (floats/integers) yo hold data from a region
temperature_data =[32.5,28.0,15.0,22.4,35.2]
humidity_data = [85,90,40,65,55]

# ---2c. Conditional statements  to determine weather conditions ---
def determine_condition(temp, humidity):
    """Determines if the weather is sunny, Rainy, etc. based on thresholds."""
    if humidity > 80:
        return "rainy / Humid"
    elif temp > 30:
        return "Sunny and Hot"
    elif temp < 18:
        return "Cold / Chilly"
    else:
        return "Pleasant / Clear"

# ---2d. Functions to calculate average and print results ---
def analyze_weather():
    # Calculate averages using standard math formulas
    avg_temp = sum(temperature_data) / len(temperature_data)
    avg_humidity = sum(humidity_data) /len(humidity_data)



    print("---Weather Data Analysis Results ---")
    print(f"Average Temparature: {avg_temp:.1f}C")
    print(f"Average Humidity   : {avg_humidity:.1f}%")
    print("-" * 38)


    # --- 2a. Display the output/condition for every individual action/reading ---
    print("Daily Weather Breakdown:")
    for i in range(len(temperature_data)):
        t = temperature_data[i]
        h = humidity_data[i]
        condition = determine_condition(t, h)
        print(f"Day {i+1}: Temp = {t}degree C,Humidity = {h}% -> Condition: {condition}")

 # Execution of the analyzer function
analyze_weather()
```
